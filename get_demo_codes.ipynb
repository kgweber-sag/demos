```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv('.env')
engine = create_engine(os.environ.get("ALAB_DB_URL"))
```

```python
# select the vocabulary_id and concept_code for ten codes in each vocabulary

codes = pd.DataFrame()

with engine.connect() as conn:
    result = conn.execute("""SELECT vocabulary_id
    FROM vocab.vocabulary
    WHERE NOT vocabulary_reference
    IN ('https://github.com/OHDSI/CommonDataModel','OMOP generated')
    AND NOT vocabulary_id = 'CMS Place of Service'""").fetchall()
    vocabularies = [row[0] for row in result]

for vocab in vocabularies:
    sql = f"""
    SELECT vocabulary_id, concept_code
    FROM vocab.concept
    WHERE vocabulary_id = '{vocab}'
    LIMIT 10
    """
    codes = pd.concat([codes, pd.read_sql(sql, engine)])
```

```python
codes = codes.loc[~codes.concept_code.str.startswith('OMOP')]

codes['stringrep'] = codes['vocabulary_id'] + ':' + codes['concept_code']
codes_list = codes['stringrep'].tolist()
```
